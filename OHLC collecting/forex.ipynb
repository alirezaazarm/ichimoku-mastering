```python
def ohlc(symbol,time_frame,n_candles,time):

    timezone = pytz.timezone("Etc/UTC")
    if time == 'now':
        utc_from = datetime.now(tz=timezone)
    else:
        utc_from = datetime.strptime(time, '%Y-%m-%d %H:%M:%S')

    if time_frame == '1m':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_M1, utc_from, n_candles)
    elif time_frame == '3m':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_M3, utc_from, n_candles)
    elif time_frame == '5m':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_M5, utc_from, n_candles)
    elif time_frame == '6m':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_M6, utc_from, n_candles)
    elif time_frame == '10m':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_M10, utc_from, n_candles)
    elif time_frame == '15m':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_M15, utc_from, n_candles)
    elif time_frame == '20m':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_M20, utc_from, n_candles)
    elif time_frame == '30m':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_M30, utc_from, n_candles)
    elif time_frame == '1h':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_H1, utc_from, n_candles)
    elif time_frame == '2h':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_H2, utc_from, n_candles)
    elif time_frame == '3h':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_H3, utc_from, n_candles)
    elif time_frame == '4h':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_H4, utc_from, n_candles)
    elif time_frame == '1d':
        rates = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_D1, utc_from, n_candles)

    rates_frame = pd.DataFrame(rates)
    rates_frame['time']=pd.to_datetime(rates_frame['time'], unit='s')
    candles = rates_frame.drop(columns=['real_volume','spread','tick_volume'])

    return candles

candles = ohlc("USDCAD_i",'5m',30000, '2023-09-03 19:30:00')
candles
```

```python
def read_from_db(title):
    engine = create_engine("mysql+pymysql://{user}:{pw}@localhost:{port}/{db}"
                       .format(user="root",
                               pw="root",
                               port=3307,
                               db="mydb"))
    sql = pd.read_sql_table(title,engine)
    df = pd.DataFrame(sql)
    return df
```

```python
import sqlalchemy as sa
def check_db(name):
    engine = create_engine("mysql+pymysql://{user}:{pw}@localhost:{port}/{db}"
                       .format(user="root",
                               pw="root",
                               port=3307,
                               db="mysql"))

    return sa.inspect(engine).has_table(name)

check_db('candles')
```
